```python
import os
import json
import shutil
from PIL import Image
image_directory = "data/image"

output_data = []




for i in range(1, 1005):
    try:
        img_path = f"{image_directory}/AMBER_{i}.jpg"

        caption_list = []
        piece_caption = run_inference(img_path, '<CAPTION>')
        caption_list.append(piece_caption)
        
        print(f"Response for image {i}: {piece_caption['<CAPTION>']}")
        output_data.append({"id": i, "response": piece_caption['<CAPTION>']})

    except Exception as e:
        print(f"Issue on ID: {i}, Error: {e}")


output_file = "generated_captions/1_florence_captions_base.json"
with open(output_file, "w") as file:
    json.dump(output_data, file, indent=4)

print(f"Evaluation results saved to {output_file}")
```
